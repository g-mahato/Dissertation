```python
X = final_df.drop('magnitude', axis = 1)
```

```python
y = final_df['magnitude']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the dataset into train test set
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
# here test set data is divided into two equal set for evaluation and test
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size = 0.5)
```

```python
len(X_train)
```

```python
len(X_eval)
```

```python
len(y_eval)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# here i will scale each data set differently to avoid any kind of data likage
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_eval = scaler.transform(X_eval)
```

## Trainning the model

```python
from sklearn.linear_model import Ridge
```

```python
# model one with alpha 100

model_one = Ridge(alpha=100)
```

```python
#trainning the model with training set
model_one.fit(X_train, y_train)
```

```python
# pridicting the evaluation set 
y_eval_pred = model_one.predict(X_eval)
```

```python
# importng the mean squared error for checking the error made by the machine learning model
from sklearn.metrics import mean_squared_error
```

```python
# the mean squared error of the evaluation set is calculatd 
mean_squared_error(y_eval, y_eval_pred)
```

```python
# model two with alpha 1
model_two = Ridge(alpha = 1)
#trainning the model with training set

model_two.fit(X_train, y_train)
```

```python
# pridicting the evaluation set 
eval_pred = model_two.predict(X_eval)
```

```python
# the mean squared error of the evaluation set is calculatd 

mean_squared_error(y_eval, eval_pred)
```

```python
np.sqrt(mean_squared_error(y_eval, y_eval_pred))
```

```python

# cross value score
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(model_two, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 5)
```

```python
score
```

```python
abs(score.mean())
```

```python
# elestivNet using gridsearch cv
```

```python
from sklearn.linear_model import ElasticNet
base_enet_model = ElasticNet(max_iter=10000000)
```

```python
# creating parameter for alpha and l1_ratio
param_grid = {'alpha':[0.1, 1, 5, 10, 50, 100],
              'l1_ratio':[0.1,0.4, 0.6,0.8, 0.95, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(
    estimator= base_enet_model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    verbose=1
)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_eval_pred = grid_model.predict(X_eval)
```

```python
# ACCURACY 
1 - mean_absolute_error(y_eval, y_eval_pred)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred)
```
